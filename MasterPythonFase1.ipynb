```python
import time

start = time.perf_counter()

def sezione1_python_internals():

    print("=" * 80)
    print("Sezione 1 python internals")
    print("=" * 80)

    #Object model
    print("Tutto Ã¨ un oggetto")
    print("=" * 80)

    x = 50.33

    print(f"x = 5000")
    print(f"ID di x = {id(x)}")
    print(f"Tipo di x = {type(x)}")
    print(f"Valore di x = {x}")
    print(f"Size in bite = {sys.getsizeof(x)}")
    print(f"Refcount = {sys.getrefcount(x)}")

    y = x
    print(f"\n y = x Cosa succede?")
    print(f"y = x ? { y is x} stesso oggetto")
    print(f"ID di x: {id(x)}")
    print(f"ID di y: {id(y)}")
    print(f"Refount ora {sys.getrefcount(x)}")

    #Immutabili 
    print("Immutabili sono: int, str, tuple")
    a = 100
    print(f" a = 100, ID : {id(a)}")
    a = a + 30.50 # Crea nuovo oggetto
    print(f" a = a + 30.50")
    print(f" a = {a}, ID = {id(a)} nuovo oggetto")

    # Mutabile
    print(f"Mutabili sono: (list, dict, set)")
    lista1 = [1, 2, 3]
    print(f"ID lista1 : {id(lista1)}")
    lista2 = lista1
    lista2.append(20.77)
    print(f"lista1 = {lista1}, ID lista1 = {id(lista1)}")
    print(f"lista2 = {lista2}, ID lista2 = {id(lista2)}")    
    
```

```python
sezione1_python_internals()

end = time.perf_counter()

print(f"Tempo trascorso = {end - start:.4f} secondi")
```

## 1. PYTHON INTERNALS

```python
# Python Object Model
import sys
import gc

x = 42
print(f"ID: {id(x)}")
print(f"Type: {type(x)}")
print(f"Value: {x}")
print(f"Size: {sys.getsizeof(x)} bytes")
print(f"Refcount: {sys.getrefcount(x) - 1}")
```

```python
# Mutable vs Immutable
# Immutable
a = 100
print(f"a = {a}, ID: {id(a)}")
a = a + 1
print(f"a = {a}, ID: {id(a)} - DIVERSO!")

# Mutable
lista = [1, 2, 3]
print(f"\nlista = {lista}, ID: {id(lista)}")
lista.append(4)
print(f"lista = {lista}, ID: {id(lista)} - STESSO!")
```

## 2. DATA TYPES

```python
# Integers - Illimitati!
huge = 10**100
print(f"10^100 has {huge.bit_length()} bits")

# Sistemi numerici
print(f"\nBinary: {0b101010}")
print(f"Octal: {0o52}")
print(f"Hex: {0x2A}")
```

```python
# Float precision problem
result = 0.1 + 0.2
print(f"0.1 + 0.2 = {result}")
print(f"Ãˆ 0.3? {result == 0.3}")

# Solution with Decimal
from decimal import Decimal
d1 = Decimal('0.1')
d2 = Decimal('0.2')
print(f"\nDecimal: {d1 + d2}")
```

```python
# Strings - Unicode
text = "Python ðŸ Master"
print(f"Text: {text}")
print(f"Length: {len(text)}")
print(f"UTF-8 bytes: {len(text.encode('utf-8'))}")

# F-strings avanzate
name = "Marco"
score = 95.456
print(f"\n{name:^20}")
print(f"{score:.2f}")
print(f"{0.856:.1%}")
```

## 3. DATA STRUCTURES

```python
# Liste - Slicing
lst = list(range(10))
print(f"List: {lst}")
print(f"lst[2:5] = {lst[2:5]}")
print(f"lst[::2] = {lst[::2]}")
print(f"lst[::-1] = {lst[::-1]}")

# List comprehension
squares = [x**2 for x in range(5)]
print(f"\nSquares: {squares}")
```

```python
# Dizionari
d = {'name': 'Marco', 'age': 30}
print(f"Dict: {d}")

# Dict comprehension
squares = {x: x**2 for x in range(5)}
print(f"Squares dict: {squares}")

# Merging (Python 3.9+)
d1 = {'a': 1}
d2 = {'b': 2}
d3 = d1 | d2
print(f"Merged: {d3}")
```

```python
# Set operations
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(f"Union: {a | b}")
print(f"Intersection: {a & b}")
print(f"Difference: {a - b}")
print(f"Symmetric: {a ^ b}")
```

## 4. CONTROL FLOW

```python
# Ternary operator
x = 10
status = "positive" if x > 0 else "non-positive"
print(f"x={x} is {status}")

# Chained comparisons
y = 5
print(f"0 <= {y} < 10? {0 <= y < 10}")

# all/any
conditions = [True, True, False]
print(f"\nall: {all(conditions)}")
print(f"any: {any(conditions)}")
```

```python
# Enumerate and zip
items = ['a', 'b', 'c']
for i, item in enumerate(items, start=1):
    print(f"{i}. {item}")

names = ['Alice', 'Bob']
ages = [25, 30]
for name, age in zip(names, ages):
    print(f"{name}: {age}")
```

## 5. FUNCTIONS

```python
# All parameter types
def full_function(pos_only, /, standard, *args, kw_only, **kwargs):
    print(f"pos_only: {pos_only}")
    print(f"standard: {standard}")
    print(f"args: {args}")
    print(f"kw_only: {kw_only}")
    print(f"kwargs: {kwargs}")

full_function(1, 2, 3, 4, kw_only=5, extra=6)
```

```python
# Closures
def outer(x):
    def inner(y):
        return x + y
    return inner

add_five = outer(5)
print(f"add_five(3) = {add_five(3)}")
```

```python
# Decorators
def timer(func):
    import time
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        print(f"Took {time.time()-start:.4f}s")
        return result
    return wrapper

@timer
def slow_func():
    import time
    time.sleep(0.1)
    return "Done"

result = slow_func()
```

```python
# Generators
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

fib = fibonacci(10)
print(list(fib))
```

## 6. GOTCHAS

```python
# Mutable default arguments - PROBLEMA
def bad_append(item, lst=[]):
    lst.append(item)
    return lst

print(bad_append(1))
print(bad_append(2))  # STESSA LISTA!

# SOLUZIONE
def good_append(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(good_append(1))
print(good_append(2))  # Liste diverse!
```

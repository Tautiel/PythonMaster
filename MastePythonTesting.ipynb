```python
import ipdb
import sys
import dis
import math
from decimal import Decimal, getcontext
from fractions import Fraction
from collections import ChainMap, namedtuple
from typing import List, Dict, Optional, Union, Callable, TypeVar
import copy
import gc
import time
```

```python
print("=" * 80)
print("SESSIONE PYTHON INTERNALS")
print("=" * 80)

def python_internals():
    ipdb.set_trace()
    """Come Pytohn gestisce la memoria, oggetti e riferimenti"""

    print("\n" + "=" * 60)
    print("SEZIONE 1: PYTHON INTERNALS")
    print("-" * 80)

    #Object model

    print("tutto è un oggetto")
    print("=" * 40)
    x = 42
    print(f"x = 42")
    print(f"ID (indirizzo memoria): {id(x)}")
    print(f"tipo: {type(x)}")
    print(f"valore: {x}")
    print(f"size in bytes: {sys.getsizeof(x)}")
    print(f"Refcount: {sys.getrefcount(x) - 1}")

    # Dimostra che è un riferimento
    
    y = x
    print(f"\n y = x Cosa succede?")
    print(f"x is y? {x is y} (stesso oggetto)")
    print(f"ID di x: {id(x)}")
    print(f"ID di y: {id(y)}")
    print(f"Refcount ora: {sys.getrefcount(x) - 1} (aumentato)")

    # Mutable vs Immutable

    print("\nDifferenza critica tra mutabile ed immutabile")
    print("=" * 60)

    # Immutabile

    print("IMMUTABILE (int, str, tuple)")
    a = 100
    print(f" a = 100:, ID: {id(a)}")
    a = a + 1 # Crea nuovo oggetto!
    print(f"a = a + 1")
    print(f" a = {a}, ID: {id(a)} <- Diverso nuovo oggetto creato")

    # Mutable

    print("\n Mutable (list, dict, set")
    lista1 = [1, 2, 3]
    print(f"lista1 = [1, 2, 3] ID: {id(lista1)}")
    lista2 = lista1 # Riferimento non copia!
    lista2.append(4)
    print(f"lista2 = lista1: {lista2.append(4)}")
    print(f"lista1 = {lista1}, ID: {id(lista1)} <- STESSO")
    print(f"lista2 = {lista2}, ID: {id(lista2)} <- STESSO")

    # Copie Shallow vs Deep

    print("\n Copie Shallow vs Deep")
    print("=" * 40)

    original = [[1, 2], [3, 4]]
    shallow = original.copy()
    deep = copy.deepcopy(original)

    print(f"original = {original}")
    original[0].append(3)
    print(f"Dopo original[0].append(3):")
    print(f"original: {original}")
    print(f"shallow: {shallow} <- inner liost modificata")
    print(f"deep: {deep} <- non modificata!")

    # Bytecode

    print("\nBytecode come python compila")
    print("=" * 40)

    def example(a, b):
        return a + b * 2
    
    print("def example(a, b):")
    print("      return a + b * 2")
    print("\n Bytecode generato: ")
    dis.dis(example)


    # Garbage collection

    print("\n Garbage collection")
    print("=" * 40)

    class Node:
        def __init__(self, value):
            self.value = value
            self.ref = None

    # Circular reference

    a = Node(1)
    b = Node(2)
    a.ref = b
    b.ref = a # Circular!

    print("Creato riferimento circolare: ")
    print("a -> b -> a")
    print(f"oggetti prima del GC: {len(gc.get_objects())}")

    del a, b # Cancella riferimenti
    collected = gc.collected() # Forza garbage collection

    print(f"oggetti raccolti dal gc: {collected}")
    print(f"oggetti dopo il gc: {len(gc.get_object())}")
```

```python
python_internals()
```
